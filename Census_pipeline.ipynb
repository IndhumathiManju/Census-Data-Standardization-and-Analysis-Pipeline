```python
import pandas as pd
# Load a sample dataset
df_P = pd.read_excel('/content/df_P')


 # Identify the rows with missing values in the "Literate" column
missing_literate = df_P['Literate'].isna()
missing_literate_Male = df_P['Literate_Male'].isna()
missing_literate_Female = df_P['Literate_Female'].isna()

# Calculate the missing values using the sum of "Male-Literate" and "Female-Literate"
df_P.loc[missing_literate, 'Literate'] = df_P['Literate_Male'] + df_P['Literate_Female']
df_P.loc[missing_literate_Male, 'Literate_Male'] = df_P['Literate'] - df_P['Literate_Female']
df_P.loc[missing_literate_Female, 'Literate_Female'] = df_P['Literate'] - df_P['Literate_Male']

#Calculating the missing data for population

missing_Population = df_P['Population'].isna()
missing_Male = df_P['Male'].isna()
missing_Female = df_P['Female'].isna()

df_P.loc[missing_Population, 'Population'] = df_P['Male'] + df_P['Female']
df_P.loc[missing_Male, 'Male'] = df_P['Population'] - df_P['Female']
df_P.loc[missing_Female, 'Female'] = df_P['Population'] - df_P['Male']

#Calculating the missing data for SC_
missing_SC = df_P['SC'].isna()
missing_Male_SC = df_P['Male_SC'].isna()
missing_Female_SC = df_P['Female_SC'].isna()

df_P.loc[missing_SC, 'SC'] = df_P['Male_SC'] + df_P['Female_SC']
df_P.loc[missing_Male_SC, 'Male_SC'] = df_P['SC'] - df_P['Female_SC']
df_P.loc[missing_Female_SC, 'Female_SC'] = df_P['SC'] - df_P['Male_SC']

#Calculating the missing data for ST_
missing_ST = df_P['ST'].isna()
missing_Male_ST = df_P['Male_ST'].isna()
missing_Female_ST = df_P['Female_ST'].isna()

df_P.loc[missing_ST, 'ST'] = df_P['Male_ST'] + df_P['Female_ST']
df_P.loc[missing_Male_ST, 'Male_ST'] = df_P['ST'] - df_P['Female_ST']
df_P.loc[missing_Female_ST, 'Female_ST'] = df_P['ST'] - df_P['Male_ST']

#Calculating the missing data for Workers
missing_Workers = df_P['Workers'].isna()
missing_Male_Workers = df_P['Male_Workers'].isna()
missing_Female_Workers = df_P['Female_Workers'].isna()

df_P.loc[missing_Workers, 'Workers'] = df_P['Male_Workers'] + df_P['Female_Workers']
df_P.loc[missing_Male_Workers, 'Male_Workers'] = df_P['Workers'] - df_P['Female_Workers']
df_P.loc[missing_Female_Workers, 'Female_Workers'] = df_P['Workers'] - df_P['Male_Workers']

# Calculationfg the missing data for Households

missing_Households = df_P['Households'].isna()
missing_Households_Rural = df_P['Households_Rural'].isna()
missing_Households_Urban = df_P['Households_Urban'].isna()

df_P.loc[missing_Households, 'Households'] = df_P['Households_Rural'] + df_P['Households_Urban']
df_P.loc[missing_Households_Rural, 'Households_Rural'] = df_P['Households'] - df_P['Households_Urban']
df_P.loc[missing_Households_Urban, 'Households_Urban'] = df_P['Households'] - df_P['Households_Rural']

# Calculationfg the missing data for Households Youth_and_Adult
# Middle_Aged
# Senior_Citizen

missing_Population = df_P['Population'].isna()
missing_Youth_and_Adult = df_P['Youth_and_Adult'].isna()
missing_Middle_Aged = df_P['Middle_Aged'].isna()
missing_Senior_Citizen = df_P['Senior_Citizen'].isna()

df_P.loc[missing_Youth_and_Adult, 'Youth_and_Adult'] = df_P['Population'] - (df_P['Middle_Aged']+  df_P['Senior_Citizen'])
df_P.loc[missing_Middle_Aged, 'Middle_Aged'] = df_P['Population'] - (df_P['Youth_and_Adult']+  df_P['Senior_Citizen'])
df_P.loc[missing_Senior_Citizen, 'Senior_Citizen'] = df_P['Population'] - (df_P['Middle_Aged']+  df_P['Youth_and_Adult'])

# Calculationfg the missing data
# Main_Workers
# Marginal_Workers
# Non_Workers
# Cultivator_Workers
# Agricultural_Workers
# Household_Workers
# Other_Workers

missing_Main_Workers = df_P['Main_Workers'].isna()
missing_Marginal_Workers = df_P['Marginal_Workers'].isna()
missing_Non_Workers = df_P['Non_Workers'].isna()
missing_Cultivator_Workers = df_P['Cultivator_Workers'].isna()
missing_Agricultural_Workers = df_P['Agricultural_Workers'].isna()
missing_Household_Workers = df_P['Household_Workers'].isna()
missing_Other_Workers = df_P['Other_Workers'].isna()

df_P.loc[missing_Main_Workers,'Main_Workers']= df_P['Population']-(df_P['Marginal_Workers']+ df_P['Non_Workers']  + df_P['Cultivator_Workers']+df_P['Agricultural_Workers'] + df_P['Household_Workers'] +df_P['Other_Workers'])
df_P.loc[missing_Marginal_Workers,'Marginal_Workers'] = df_P['Population']-(df_P['Main_Workers']+ df_P['Non_Workers'] +df_P['Cultivator_Workers']+df_P['Agricultural_Workers'] +df_P['Household_Workers'] +df_P['Other_Workers'])
df_P.loc[missing_Non_Workers,'Non_Workers'] = df_P['Population']-(df_P['Marginal_Workers'] + df_P['Main_Workers'] +df_P['Cultivator_Workers']+df_P['Agricultural_Workers'] +df_P['Household_Workers'] +df_P['Other_Workers'])
df_P.loc[missing_Cultivator_Workers,'Cultivator_Workers'] = df_P['Population']-(df_P['Marginal_Workers']+ df_P['Main_Workers'] +df_P['Non_Workers'] +df_P['Agricultural_Workers'] +df_P['Household_Workers'] +df_P['Other_Workers'])
df_P.loc[missing_Agricultural_Workers,'Agricultural_Workers'] = df_P['Population']-(df_P['Marginal_Workers']+ df_P['Main_Workers'] +df_P['Non_Workers'] +df_P['Cultivator_Workers']+df_P['Household_Workers'] +df_P['Other_Workers'])
df_P.loc[missing_Household_Workers,'Household_Workers'] = df_P['Population']-(df_P['Marginal_Workers']+ df_P['Main_Workers'] +df_P['Non_Workers'] +df_P['Cultivator_Workers']+df_P['Agricultural_Workers'] +df_P['Other_Workers'])
df_P.loc[missing_Other_Workers,'Other_Workers'] = df_P['Population']-( df_P['Main_Workers'] +df_P['Marginal_Workers']+ df_P['Non_Workers'] + df_P['Cultivator_Workers']+df_P['Agricultural_Workers'] + df_P['Household_Workers'] )


# Calculationfg the missing data
# Hindus
# Muslims
# Christians
# Sikhs
# Buddhists
# Jains
# Others_Religions
# Religion_Not_Stated

missing_Hindus = df_P['Hindus'].isna()
missing_Muslims = df_P['Muslims'].isna()
missing_Christians = df_P['Christians'].isna()
missing_Sikhs = df_P['Sikhs'].isna()
missing_Buddhists = df_P['Buddhists'].isna()
missing_Jains = df_P['Jains'].isna()
missing_Others_Religions = df_P['Others_Religions'].isna()
missing_Religion_Not_Stated = df_P['Religion_Not_Stated'].isna()

df_P.loc[missing_Hindus,'Hindus']=df_P['Population']-( df_P['Muslims'] + df_P['Christians'] + df_P['Sikhs'] + df_P['Buddhists'] + df['Jains'] +df_P['Others_Religions'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Muslims,'Muslims'] = df_P['Population']- (df_P['Hindus'] + df_P['Christians'] + df_P['Sikhs'] + df_P['Buddhists'] + df['Jains'] +df_P['Others_Religions'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Christians,'Christians'] = df_P['Population']-(df_P['Hindus'] + df_P['Muslims'] + df_P['Sikhs'] + df_P['Buddhists'] + df['Jains'] +df_P['Others_Religions'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Sikhs,'Sikhs'] = df_P['Population']- (df_P['Hindus'] + df_P['Muslims'] + df_P['Christians'] + df_P['Buddhists'] + df['Jains'] +df_P['Others_Religions'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Buddhists,'Buddhists'] = df_P['Population']- (df_P['Hindus'] + df_P['Muslims'] + df_P['Christians'] +df_P['Sikhs'] + df['Jains'] +df_P['Others_Religions'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Jains,'Jains'] = df_P['Population']- (df_P['Hindus'] + df_P['Muslims'] + df_P['Christians'] +df_P['Sikhs'] +df_P['Buddhists']  +df_P['Others_Religions'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Others_Religions,'Others_Religions'] = df_P['Population']-(df_P['Hindus'] + df_P['Muslims'] + df_P['Christians'] +df_P['Sikhs'] +df_P['Buddhists']+ df['Jains'] + df_P['Religion_Not_Stated'])
df_P.loc[missing_Religion_Not_Stated,'Religion_Not_Stated'] = df_P['Population']- (df_P['Hindus'] + df_P['Muslims'] + df_P['Christians'] +df_P['Sikhs'] +df_P['Buddhists'] + df['Jains'] +df_P['Others_Religions'])




df_P.to_excel('df_R2', engine = 'xlsxwriter')
```

Task 4. evaluate_missing_data before and after data cleaning

```python
import pandas as pd
from prettytable import PrettyTable

def calculate_missing_data_percentage(df):
    total_cells = df.size
    missing_cells = df.isnull().sum().sum()
    missing_percentage = (missing_cells / total_cells) * 100
    return missing_percentage

def calculate_column_missing_percentage(df):
    missing_percentage_per_column = df.isnull().sum() / len(df) * 100
    return missing_percentage_per_column

def evaluate_missing_data(file_path1, file_path2):
    # Read CSV files into DataFrames
    df1 = pd.read_excel('/content/df_P')
    df2 = pd.read_excel('/content/df_H')

    # Calculate overall missing data percentage
    missing_percentage_before_data_Cleaning = calculate_missing_data_percentage(df1)
    missing_percentage_after_data_Cleaning = calculate_missing_data_percentage(df2)

    # Calculate missing data percentage per column
    column_missing_percentage_before_data_Cleaning = calculate_column_missing_percentage(df1)
    column_missing_percentage_after_data_Cleaning = calculate_column_missing_percentage(df2)

    # Create tables for displaying the results
    overall_table = PrettyTable()
    overall_table.field_names = ["File", "Overall Missing Data Percentage"]
    overall_table.add_row([file_path1, f"{missing_percentage_before_data_Cleaning:.2f}%"])
    overall_table.add_row([file_path2, f"{ missing_percentage_after_data_Cleaning:.2f}%"])

    column_table1 = PrettyTable()
    column_table1.field_names = ["Column", "Missing Data Percentage"]
    for column, percentage in column_missing_percentage_before_data_Cleaning.items():
        column_table1.add_row([column, f"{percentage:.2f}%"])

    column_table2 = PrettyTable()
    column_table2.field_names = ["Column", "Missing Data Percentage"]
    for column, percentage in column_missing_percentage_after_data_Cleaning.items():
        column_table2.add_row([column, f"{percentage:.2f}%"])

    print(f"Overall Missing Data Percentage:\n{overall_table}")
    print(f"\nMissing Data Percentage Per Column for  before data cleaning:\n{column_table1}")
    print(f"\nMissing Data Percentage Per Column for after data cleaning:\n{column_table2}")

# Paths to the CSV files
file_path1 = 'Before_cleaning'
file_path2 = 'After_Cleaning' # Replace with the path to your second CSV file

# Evaluate missing data
evaluate_missing_data(file_path1, file_path2)

#  # Create DataFrame for results
# results = pd.DataFrame({
#         'Column': df.columns,
#         'Missing Percentage Before Cleaning': column_table1.values,
#         'Missing Percentage After Cleaning': column_table2.values
#     })
# # overall_results = pd.DataFrame({
# #         'Overall Missing Percentage': [  missing_percentage_before_data_Cleaning,missing_percentage_after_data_Cleaning ],
# #         'Description': ['Before Cleaning', 'After Cleaning']
# #     })

#     # Write the results to CSV files
# results.to_csv('missing_data_percentage_per_column.csv', index=False)
# overall_results.to_csv('overall_missing_data_percentage.csv', index=False)

# print("Missing data percentage before and after cleaning has been saved to CSV files.")



```
